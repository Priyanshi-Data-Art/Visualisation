5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.


````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(0)

# Generate synthetic dataset
num_samples = 100
# Generate correlated features
x = np.random.rand(num_samples)
y = 2 * x + np.random.normal(0, 0.1, num_samples)  # y is correlated with x
z = 0.5 * x + np.random.normal(0, 0.2, num_samples)  # z is also correlated with x

# Create a DataFrame
data = pd.DataFrame({'Feature X': x, 'Feature Y': y, 'Feature Z': z})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Set up the matplotlib figure
plt.figure(figsize=(8, 6))

# Create a heatmap of the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar_kws={"shrink": .8})

# Add titles and labels
plt.title('Correlation Matrix Heatmap')
plt.show()
````

# **2. Heatmap for Sales Data**

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(20)

# Generate the sales dataset
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data for the heatmap
heatmap_data = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='sum')

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.0f')
plt.title('Heatmap of Sales Across Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()
````

```
x = np.linspace(-5, 5, 100)
```

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate z data
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame (optional for further use)
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Add color bar
fig.colorbar(surf)

# Set labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

# Show the plot
plt.show()
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(25)

# Create the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))
bubble = plt.scatter(df['GDP'], df['Population'],
                     s=df['Population'] * 2,  # Bubble size proportional to population
                     alpha=0.5,
                     edgecolors="w",
                     linewidth=2)

# Add labels and title
plt.title('Bubble Chart: Population vs GDP by Country')
plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')

# Add country labels next to each bubble
for i in range(len(df)):
    plt.text(df['GDP'][i], df['Population'][i],
             df['Country'][i],
             horizontalalignment='center',
             verticalalignment='center')

# Show the plot
plt.grid(True)
plt.show()
````

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

````python
# Import necessary libraries
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the data
x = np.linspace(0, 10, 100)  # Generate 100 points from 0 to 10
y = np.sin(x)  # Compute the sine of each x value

# Output to Jupyter Notebook
output_notebook()

# Create a Bokeh plot
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='Sin(X)',
           width=800, height=400)  # Use width and height instead

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="Sine Wave", line_width=2, color='blue')

# Show the plot
show(p)
````

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

````python
# Import necessary libraries
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Prepare the data
np.random.seed(42)  # For reproducibility
n = 50  # Number of points
x = np.random.rand(n) * 100  # Random x values between 0 and 100
y = np.random.rand(n) * 100  # Random y values between 0 and 100
sizes = np.random.randint(5, 50, n)  # Random sizes for markers between 5 and 50
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], n)  # Random colors

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Output to Jupyter Notebook
output_notebook()

# Create a Bokeh scatter plot
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y',
           width=800, height=400)

# Add a scatter renderer
p.scatter('x', 'y', size='sizes', color='colors', source=source, alpha=0.6)

# Show the plot
show(p)
````

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```

```

````python
# Import necessary libraries
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Prepare the data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Output to Jupyter Notebook
output_notebook()

# Create a Bokeh bar chart
p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label='Fruits', y_axis_label='Counts')

# Add bars to the plot
p.vbar(x='fruits', top='counts', source=source, width=0.4,
        color='skyblue', alpha=0.7)

# Show the plot
show(p)

````

````python
# Import necessary libraries
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Prepare data for Bokeh
x = edges[:-1]  # left edges of the bins
y = hist        # heights of the bins

# Output to Jupyter Notebook
output_notebook()

# Create a Bokeh histogram
p = figure(title="Histogram of Random Data",
           x_axis_label='Value',
           y_axis_label='Frequency',
           tools="reset")

# Add bars to the plot
p.quad(top=y, bottom=0, left=x, right=edges[1:], fill_color='skyblue', alpha=0.7)

# Show the plot
show(p)

````

5. Create a Bokeh heatmap using the provided dataset.
```
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
```

````python
# Import necessary libraries
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource
from bokeh.layouts import gridplot

# Prepare the data
data_heatmap = np.random.rand(10, 10)  # Random 10x10 data
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Prepare data for Bokeh
source = ColumnDataSource(data=dict(
    x=xx.flatten(),
    y=yy.flatten(),
    value=data_heatmap.flatten()
))

# Create a color mapper
mapper = linear_cmap(field_name='value', palette='Viridis256', low=data_heatmap.min(), high=data_heatmap.max())

# Create the figure
p = figure(title="Heatmap Example",
           x_axis_label='X Axis',
           y_axis_label='Y Axis',
           tools="")

# Create the heatmap with rectangles
p.rect(x='x', y='y', width=0.1, height=0.1, source=source, line_color=None, fill_color=mapper)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], location=(0, 0))
p.add_layout(color_bar, 'right')

# Output to Jupyter Notebook
output_notebook()

# Show the plot
show(p)
````
